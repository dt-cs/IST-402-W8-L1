```python
# =====================================================
#  BLIP Image Caption Generator â€” Colab Version
# =====================================================
# ==================== STEP 1: Clean Environment ====================
print("ðŸ§¹ Cleaning up...")
import os
os.system('pkill -9 streamlit')
os.system('pkill -9 ngrok')

# Install dependencies
!pip install streamlit transformers pillow pyngrok -q

# Import libraries
import os
import streamlit as st
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
!pip install pyngrok -q
from pyngrok import ngrok
ngrok.set_auth_token("NGROK TOKEN") # put your auth token here

!ngrok api tunnel-sessions stop --all

#Start ngrok tunnel (to access Streamlit UI)
public_url = ngrok.connect(8501)
print("Streamlit app running at:", public_url)

#Save Streamlit app as a Python script
with open("app.py", "w") as f:
    f.write("""
import streamlit as st
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image

@st.cache_resource
def load_model():
    processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
    model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")
    return processor, model

processor, model = load_model()

st.title("ðŸ–¼ï¸ Image to Caption Generator (BLIP Model)")
uploaded_file = st.file_uploader("Upload an Image", type=["jpg", "jpeg", "png"])

if uploaded_file is not None:
    image = Image.open(uploaded_file).convert("RGB")
    st.image(image, caption="Uploaded Image", use_column_width=True)
    st.write("Generating caption...")

    inputs = processor(image, return_tensors="pt")
    out = model.generate(**inputs)
    caption = processor.decode(out[0], skip_special_tokens=True)

    st.subheader("ðŸ“ Generated Caption:")
    st.write(caption)
""")

#Run the Streamlit app
!streamlit run app.py &>/dev/null&
```
